The age column probably won't be necessary for our model, and use dropna in this situation isn't viable because that will be a important lost data, so let's fill the NaN values with the mean of all other values

```python
train_dt.Age.fillna(train_dt.Age.mean(), inplace=True)
```

```python
train_dt.info()
```

```python
train_dt.Age = train_dt.Age.astype(int)
```

```python
train_dt.info()
```

```python
train_dt[train_dt.Survived == 1]
```

Less than 50% of passengers survived, let's check if exists relation with the passanger's gender

```python
train_dt.Sex
```

```python
train_dt.info()
```

```python
train_dt.Sex = train_dt.Sex.astype("string")
```

```python
train_dt.Sex
```

```python
train_dt.Sex[train_dt.Sex == "female"]
```

```python
train_dt.Sex[(train_dt.Sex == "female") & (train_dt.Survived == 0)]
```

Here we can view that the most part of women survived

```python
train_dt.Sex[train_dt.Sex == "male"]
```

```python
train_dt.Sex[(train_dt.Sex == "male") & (train_dt.Survived == 0)]
```

Here, we have that the major part of the men died and we discover that most men that women died too!
